## Q1

**1.**

- **Simple Linear Regression** involves one independent variable predicting one dependent variable, with the model $ y = b_0 + b_1x + \epsilon $.

- **Multiple Linear Regression** uses multiple independent variables to predict one dependent variable, modeled as $ y = b_0 + b_1x_1 + b_2x_2 + \dots + b_nx_n + \epsilon $.

- **Benefits of Multiple Linear Regression:**
  1. **More Information**: Accounts for multiple factors affecting the outcome, improving accuracy.
  2. **Enhanced Predictive Power**: Typically reduces error, leading to better predictions.
  3. **Confounding Control**: Helps isolate each variable’s unique impact.
  4. **Greater Flexibility**: Suits complex, real-world scenarios where multiple variables are at play.

**2.** 
- **Continuous Variable**:$ b_1 $ shows a rate of change.
- **Indicator Variable**: $ b_1 $ shows a group difference or shift in level.

**3.**

- When a single **indicator variable** is added alongside a **continuous variable** in **Multiple Linear Regression**, the model adjusts to capture **two separate linear relationships** for each group defined by the indicator variable. 

- **Model Behavior Change**
  1. **Separate Slopes**: The indicator variable allows the model to estimate a unique intercept for each group (e.g., male vs. female) while sharing the same slope for the continuous variable. This approach enables the model to shift up or down based on group differences while still capturing the overall trend in the data.

  2. **Two Linear Forms**:
   - **Simple Linear Regression (SLR)**: $ y = b_0 + b_1x + \epsilon\ $  
     In SLR, only the continuous variable $ x $ predicts $ y $, assuming a single linear trend for all data points.
   - **Multiple Linear Regression (MLR)** with Indicator: $ y = b_0 + b_1x + b_2d + \epsilon$  
     Here, $ d $ is the indicator variable, adding a unique intercept for each group (when $ d = 0 $ vs. $ d = 1 $), allowing distinct linear trends for each category, which better fits data with inherent group differences.
     
**4.**

- Adding an **interaction term** between a continuous variable and an indicator variable in **Multiple Linear Regression** allows the model to fit **different slopes** for each group defined by the indicator variable. This interaction term lets the effect of the continuous variable vary between the groups, giving each group its own unique line in the model.

- **Linear Form with Interaction**
The model with an interaction term is:
$$
y = b_0 + b_1x + b_2d + b_3(x \cdot d) + \epsilon
$$
where:
- $ x $ is the continuous variable,
- $ d $ is the indicator variable,
- $ x \cdot d $ is the interaction term.

- **Effect**
- **Separate Slopes and Intercepts**: The term $ b_3(x \cdot d) $ allows the slope of $ x $ to differ by group, meaning both the intercepts and slopes vary between groups. This makes the model more flexible and accurate in cases where the relationship between $ x $ and $ y $ changes depending on the category.

**5.**

- **Binary Encodings**: Each indicator variable $ d_i $ is set to 1 if the observation belongs to category $ i $ and 0 otherwise.
- **Reference Category**: The intercept $ b_0 $ represents the mean outcome for the reference category (the category without a dummy variable), while each coefficient $ b_i $ shows the difference in the mean outcome between category $ i $ and the reference.

## Q5

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

```python
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

## Q8

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Define a list to store performance metrics
in_sample_metrics = []
out_of_sample_metrics = []

# Loop through 10 iterations
for i in range(10):
    # Split the data into training and testing sets without using np.random.seed
    train_data, test_data = train_test_split(pokeaman, test_size=0.2)
    
    # Define the model
    model_spec = smf.ols('HP ~ Attack + Defense + Speed + Q("Sp. Def")', data=train_data)
    model_fit = model_spec.fit()
    
    # In-sample predictions
    in_sample_preds = model_fit.predict(train_data)
    in_sample_rmse = np.sqrt(mean_squared_error(train_data['HP'], in_sample_preds))
    in_sample_r2 = r2_score(train_data['HP'], in_sample_preds)
    in_sample_metrics.append((in_sample_rmse, in_sample_r2))
    
    # Out-of-sample predictions
    out_of_sample_preds = model_fit.predict(test_data)
    out_of_sample_rmse = np.sqrt(mean_squared_error(test_data['HP'], out_of_sample_preds))
    out_of_sample_r2 = r2_score(test_data['HP'], out_of_sample_preds)
    out_of_sample_metrics.append((out_of_sample_rmse, out_of_sample_r2))

# Convert results into dataframes for better visualization
in_sample_df = pd.DataFrame(in_sample_metrics, columns=['RMSE', 'R2'])
out_of_sample_df = pd.DataFrame(out_of_sample_metrics, columns=['RMSE', 'R2'])

# Visualize the performance metrics
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot RMSE
ax[0].plot(in_sample_df['RMSE'], label='In-Sample RMSE', marker='o')
ax[0].plot(out_of_sample_df['RMSE'], label='Out-of-Sample RMSE', marker='o')
ax[0].set_title('RMSE Comparison')
ax[0].legend()

# Plot R2
ax[1].plot(in_sample_df['R2'], label='In-Sample R2', marker='o')
ax[1].plot(out_of_sample_df['R2'], label='Out-of-Sample R2', marker='o')
ax[1].set_title('R2 Comparison')
ax[1].legend()

plt.tight_layout()
plt.show()

```
