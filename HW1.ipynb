## Q4

Here’s how to understand the discrepancies you might observe between `df.shape` and the summary reported by `df.describe()`:

### (a) Number of Columns

- **`df.shape`**: This gives you the dimensions of the dataset, indicating the total number of rows and columns. For example, if `df.shape` is `(100, 5)`, it means there are 100 rows and 5 columns.

- **`df.describe()`**: This function provides a summary of the dataset, but it doesn’t necessarily include all columns, especially if some columns are of non-numeric data types. By default, `df.describe()` summarizes only numeric columns. To include categorical data, you need to use `df.describe(include='all')`.

**Discrepancy Explanation**: If your dataset has non-numeric columns and you use `df.describe()` without specifying `include='all'`, it will only show statistics for numeric columns. This could lead to fewer columns being reported in the summary compared to what `df.shape` shows.

### (b) Values Reported in the "Count" Column

- **`df.shape`**: Provides the total number of rows in the dataset. The count of non-null values in each column reported by `df.describe()` should ideally match this number unless there are missing values.

- **`df.describe()`**: The "count" column in the summary indicates the number of non-null entries for each column. If there are missing values in any column, this count will be less than the total number of rows. 

**Discrepancy Explanation**: The discrepancy between the total number of rows (`df.shape[0]`) and the count of non-null values reported in `df.describe()` for each column occurs because some columns may have missing values. If a column has missing values, its count will be less than the total number of rows.

To sum up, the key points are:
- `df.shape` shows the overall size of the dataset, including all columns.
- `df.describe()` provides a summary of columns but may exclude non-numeric columns by default, leading to potential discrepancies in the number of columns reported.
- Missing values in the dataset will cause the "count" in `df.describe()` to be less than the total number of rows.
